```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
no_aug_df = pd.read_csv('../data/processed/phi_no_aug/train.csv')
flip_df = pd.read_csv('../data/processed/phi_flip/train.csv')
```

```python
plt.plot(no_aug_df['accuracy'], label='no_augmentation accuracy')
plt.plot(flip_df['accuracy'], label='flip accuracy')
plt.legend()
```

```python
noaug_epoch_df = no_aug_df.groupby('epoch')
flip_epoch_df = flip_df.groupby('epoch')
plt.plot(noaug_epoch_df.mean()['loss'], label='no_augmentation  accuracy')
plt.plot(flip_epoch_df.mean()['loss'], label='flip accuracy')
plt.legend()
```
