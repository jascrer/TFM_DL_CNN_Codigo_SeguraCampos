```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
paperlr_df = pd.read_csv('../data/processed/phi_no_aug/test.csv')
gdlr_df = pd.read_csv('../data/processed/phi_flip/test.csv')
# smallestlr_df = pd.read_csv('../data/processed/alex_flip/test.csv')
```

```python
plt.plot(paperlr_df['accuracy'], label='paper lr accuracy')
plt.plot(gdlr_df['accuracy'], label='gdlr accuracy')
# plt.plot(smallestlr_df['accuracy'], label='smallest accuracy')
plt.legend()
```

```python
paperlr_epoch_df = paperlr_df.groupby('epoch')
gdlr_epoch_df = gdlr_df.groupby('epoch')
#smallestlr_epoch_df = smallestlr_df.groupby('epoch')
plt.plot(paperlr_epoch_df.mean()['loss'], label='paperlr loss')
plt.plot(gdlr_epoch_df.mean()['loss'], label='gdlr loss')
#plt.plot(smallestlr_epoch_df.mean()['loss'], label='smallestlr loss')
plt.legend()
```
