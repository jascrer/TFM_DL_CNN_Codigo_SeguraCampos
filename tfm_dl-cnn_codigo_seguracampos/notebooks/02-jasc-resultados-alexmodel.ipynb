# Sección para importar librerías

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_model_dataset(location: str, group_by_field='epoch'):
    model_info = pd.read_csv(location)
    group_by_model = model_info.groupby(group_by_field)
    return group_by_model
```

# Resultados del entrenamiento para PhiLeNet Model

```python
no_aug_training_model_paperlr = load_model_dataset('../data/processed/alex_no_aug/paperlr/train.csv')
no_aug_testing_model_paperlr = load_model_dataset('../data/processed/alex_no_aug/paperlr/test.csv')
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
fig.suptitle('Training and Testing Data AlexLeNet Model With No Augmentations with LR = 0.01', fontsize=16)
ax[0].set_title("Accuracy Metrics")
ax[0].plot(no_aug_training_model_paperlr.mean()['accuracy'], label='Mean accuracy in the training model')
ax[0].plot(no_aug_testing_model_paperlr.mean()['accuracy'], label='Mean accuracy in the testing model')
ax[0].legend()
ax[1].set_title("Loss Metrics")
ax[1].plot(no_aug_training_model_paperlr.mean()['loss'], label='Mean loss in the training model')
ax[1].plot(no_aug_testing_model_paperlr.mean()['loss'], label='Mean loss in the testing model')
ax[1].legend()
```
